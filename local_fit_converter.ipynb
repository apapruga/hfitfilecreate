# Local CSV to FIT Converter

Этот notebook позволяет локально запускать конвертацию CSV в FIT без Telegram-бота.

Поддерживаются два сценария:
- конвертация из CSV-файла
- конвертация из CSV-текста прямо в notebook

```python
from pathlib import Path

from fit_builder import convert_csv_to_fit, create_sample_csv
```

```python
PROJECT_DIR = Path.cwd()
OUTPUT_DIR = PROJECT_DIR / "notebook_output"
OUTPUT_DIR.mkdir(exist_ok=True)

print("PROJECT_DIR:", PROJECT_DIR)
print("OUTPUT_DIR:", OUTPUT_DIR)
```

## Вариант 1. Создать пример CSV и сконвертировать его

```python
sample_csv_path = OUTPUT_DIR / "sample_workout.csv"
sample_fit_path = OUTPUT_DIR / "sample_workout.fit"

create_sample_csv(sample_csv_path)
convert_csv_to_fit(sample_csv_path, sample_fit_path, workout_name="Sample Workout")

print("CSV:", sample_csv_path)
print("FIT:", sample_fit_path)
```

## Вариант 2. Вставить CSV текстом и сконвертировать

```python
csv_text = """step_name,step_type,duration_sec,pace_min,pace_max,avg_pace,hr_min,hr_max,avg_hr,repeats
Разминка,warmup,720,7:20,7:35,,,,,1
Ускорение,interval,20,,,,165,175,,4
Восстановление,recovery,40,,,,130,145,,4
Основной интервал,interval,120,6:00,6:10,,,,,6
Восстановление,recovery,120,,,,,,140,6
Заминка,cooldown,600,7:20,7:50,,,,,1
"""

text_csv_path = OUTPUT_DIR / "workout_from_text.csv"
text_fit_path = OUTPUT_DIR / "workout_from_text.fit"

text_csv_path.write_text(csv_text, encoding="utf-8")
convert_csv_to_fit(text_csv_path, text_fit_path, workout_name="Workout From Text")

print("CSV:", text_csv_path)
print("FIT:", text_fit_path)
```

## Вариант 3. Использовать свой CSV-файл

```python
# Укажи путь к своему CSV-файлу и имя результата.
# Пример:
# user_csv_path = Path("/absolute/path/to/your/workout.csv")

user_csv_path = OUTPUT_DIR / "sample_workout.csv"
user_fit_path = OUTPUT_DIR / "my_workout.fit"
workout_name = "My Workout"

convert_csv_to_fit(user_csv_path, user_fit_path, workout_name=workout_name)

print("CSV:", user_csv_path)
print("FIT:", user_fit_path)
```
